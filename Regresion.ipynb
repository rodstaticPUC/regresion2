# Regresion: Diabetes

Conjunto de datos con una dimensionalidad de 10 variables y 442 obs.

```python

from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

diabetes = load_diabetes()
print(diabetes.DESCR) # ofrece una descripcion de la data
```

```python
X = diabetes.data # conjunto de datos
y = diabetes.target # conjunto de datos objetivo
AGE = 0; SEX = 1; MASS_INDEX = 2; BLOOD_PRESSURE = 3;


print(X.shape) # Dimewnsion de la data, 442 obs y 10 variables

print(y.shape) #variable adecuada para la regresion, ya que es numerica

print(X[0:5,:] ) # muestro las primeras filas de la data

print(y[0:5] ) # muestro las primeras filas de la data objetivo

print(X[1, MASS_INDEX] ) # muestro la característica de MASS_INDEX de la tercera fila

print(X[1:3, AGE] ) # muestro la característica de AGE de la primera y segunda fila

print(diabetes.data.shape[0] )
```

```python
x = range(0, diabetes.data.shape[0],1) # ambos caminos son iguales
x = range(0, len(y),1)
x
```

# Graficos

```python
plt.figure(figsize = (30,8)) # dimensiones del grafico
plt.plot(x,X[:, MASS_INDEX], 'y-', label='Indice de Masa')
plt.plot(x,X[:, BLOOD_PRESSURE], 'b-', label='Presion Sanguinea')
plt.rcParams.update({'font.size':22});
plt.legend(prop={'size':25});
plt.grid(),
plt.show()
```

# Diagrama de dispersion
* 1.- Indice de masa vs presion arterial
* 2.- Edad vs presion arterial

```python
fig, axs = plt.subplots(1,2, figsize = (22,6))
for fig, feature, label in zip(range(2),[MASS_INDEX,AGE],['Indice de Masa','Edad']):
     axs[fig].scatter(X[:,feature],X[:,BLOOD_PRESSURE],c='r',marker='o')
     axs[fig].set_xlabel(label);
     axs[fig].set_ylabel('Presion Arterial');
     axs[fig].grid();
plt.show()
```
